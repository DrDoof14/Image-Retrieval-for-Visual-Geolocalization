## 2.1.visualize dataset

```python
def vis_data_set(dataset_name,ty ,sample_images,project_directory =project_directory,save_location=None, file_name=None):
  fig, axs = plt.subplots(1,10,figsize=(40,8))
  plt.tight_layout()
  random_list = []

  for i in range(10):
    
      random_list.append(random.randint(0, len(sample_images)))
  index = 1
  for c in random_list:    
    img_add = os.path.join('/content/drive/MyDrive/MLDL/datasets_vg/',dataset_name, 'images/train',ty,sample_images[c] )    
    img = cv2.imread( img_add)
    img =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  
    plt.subplot(1,10, index)   
    index+=1        
    plt.imshow(img, cmap="gray",interpolation='none')
  if file_name:
      path = os.path.join(save_location, file_name+'.jpg') 
      print('sample images have been saved in: ',path)
      plt.savefig(path)
```
